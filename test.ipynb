```python
import os
import time

def clear_terminal():
    os.system('clear')  # Linux清空终端的命令

def print_clock(hour, minute, second):
    # print("================================", end='',flush=True)
    print(f"\r{hour:02d}:{minute:02d}:{second:02d}", end='', flush=True)
    
    print("\n================================", end='',flush=True)
    

def main():
    try:
        while True:
            current_time = time.localtime()
            hour, minute, second = current_time.tm_hour, current_time.tm_min, current_time.tm_sec

            clear_terminal()
            print_clock(hour, minute, second)

            time.sleep(1)
    except KeyboardInterrupt:
        print("\n退出循环，程序结束。")

if __name__ == "__main__":
    main()
```
