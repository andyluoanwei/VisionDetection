```python
import cv2 as cv
from matplotlib import pyplot as plt
from matplotlib.widgets import Slider
import numpy as np

class ImageProcessor:
    def __init__(self, image_path=None, image_flags=None):
        
        pass
        
        
    # Define the image reading function
    def read_image(self,image_path, image_flags):
        """# ReadImage

        Args:
            image_path (_type_): str,imae file path
            key (_type_): Image flag

        Returns:
            _type_: numpy.ndarray
        """
        
        image = cv.imread(image_path, image_flags)
        
        
        return image
    
    
    def CaptureVideo(image_path):
        
        pass
    
    def save_imgae(self, image_name, image_path):
        """# SaveImage

        Args:
            image_path (_type_): str,imae file path
            image_name (_type_): str,image name

        Args:
            image_path (_type_): _description_
            image_name (_type_): _description_
        """
        
        cv.imwrite(image_path,image_name)
        
        return "Save image file OK!"


    def show_images_with_title(self, images, titles):
        """Show images with title

        Args:
            images (_type_): _description_
            titles (_type_): _description_
        """
        
        num_images = len(images)                            # 
        num_rows = int(np.ceil(np.sqrt(num_images)))        # num_images
                                                            # np.sqrt(num_images)
                                                            # np.ceil()
                                                            # int()
        
        num_clos = int(np.ceil(num_images / num_rows))      # num_rows
                                                            # num_images / num_rows
                                                            # np.ceil()
                                                            # int()
                                                            
        fig, axs = plt.subplots(num_rows, num_clos, figsize=(10, 8))    # plt.subplots(num_rows, num_cols, figsize=(10, 8))num_rowsnum_cols
                                                                        # figsize=(10, 8)10x8    
                                                                        # self.figMatplotlibFigure
                                                                        # self.axsself.axs[row, col]
        for i in range(num_images):
            image = images[i]
            rgb_imaeg = cv.cvtColor(image, cv.COLOR_BGR2RGB)
            row = i // num_clos
            col = i % num_clos
            axs[row, col].imshow(rgb_imaeg)
            axs[row, col].set_title(titles[i])
            axs[row, col].axis('off')
            
        plt.tight_layout()
        plt.axis("off")
        
        plt.show()


    def run(self):
        """
        # Run the code\n
        Color:\tcv2.IMREAD_COLOR1 \n
        Gray:\tcv2.IMREAD_GRAYSCALE0 \n
        Alpha:\tcv2.IMREAD_UNCHANGEDalpha-1
        """
        image = []
        
        Image_Flags = {
            "-1": cv.IMREAD_UNCHANGED,
            "0": cv.IMREAD_GRAYSCALE,
            "1": cv.IMREAD_COLOR,
        }

        rgb_image = self.read_image("./image.jpg", Image_Flags["1"])
        gray_image = self.read_image("./image.jpg", Image_Flags["0"])
        alpha_image = self.read_image("./image.jpg", Image_Flags["-1"])
        hsv_image = cv.cvtColor(rgb_image, cv.COLOR_BGR2HSV)
        
        image = [rgb_image, gray_image, alpha_image, hsv_image]
        title = ["rgb_image", "gray_image", "alpha_image", "hsv_image"]
        
        self.show_images_with_title(image, title)        
        

if __name__ == "__main__":
    # Use the default configuration file
    start = ImageProcessor
    start().run()
```
